```python
# MDRO's & 微生物報告

mdro_1 = mdro_1[['病歷號碼', 'time', '就醫序號', '申請編號', '菌株', '預測值', '部位(所有部位)', '資訊室資料發送時間']]
mdro_1.rename({'time':'openTime', '部位(所有部位)':'部位', '資訊室資料發送時間':'sendTime'}, axis=1, inplace=True)


```

```python
report_1 = report_1[['病歷號碼', 'bed', '就醫序號', '申請編號', '部位', '抗藥性菌株', 'reportTime(發報告時間)']]
report_1.rename({'reportTime(發報告時間)':'reportTime'}, axis=1, inplace=True)
report_1
```

```python
mdro_2 = mdro_2[['病歷號碼', 'time', '就醫序號','申請編號', '菌株', '預測值', '部位', 'sendTime']]
mdro_2.rename({'time':'openTime'}, axis=1, inplace=True)
mdro_2
```

```python
report_2 = report_2[['病歷號碼', 'bed', '就醫序號','申請編號', '部位', '抗藥性菌株', 'reportTime']]
report_2
```

```python
cdss_1 = pd.merge(report_1, mdro_1, how='left', on=['病歷號碼', '就醫序號','申請編號', '部位'])
cdss_2 = pd.merge(report_2, mdro_2, how='left', on=['病歷號碼', '就醫序號','申請編號', '部位'])
# cdss_2.info()
```

```python
cdss_1 = cdss_1.drop_duplicates()
cdss_2 = cdss_2.drop_duplicates()
```

```python
cdss_2.info()
```

```python
CDSS = pd.concat([cdss_1, cdss_2],axis=0)
CDSS = CDSS.drop_duplicates()

# 如同一菌株預測多筆，則選第一筆出現
CDSS['minTime'] = CDSS.groupby(['病歷號碼', 'bed','就醫序號', '申請編號', '部位', '菌株', '預測值',  'sendTime'])['openTime'].transform('min')
```
